## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_________________________________________________________________________________________

Primary Keys are unique to each row of a given table. 
Foreign keys are primary in another table, but may not be unique to a given row in the current table.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_________________________________________________________________________________________

Default/Inner Join drops rows that do not exist in both tables.
Left Join keeps all of the original rows and adds to them from the new table, but drops rows in the new table that did not relate to the original table. 
Right Join does the opposite, keeping all rows from the new table.
Full Join combines all rows from both tables
Cross Join combines each row of table 1 with each row of table 2; it's a cartesian join.

## Question 6

What role do Associative Entities play (JOIN Tables) in many-to-many JOINs?

Type your answer below this line:
_________________________________________________________________________________________

An associative entity is a table created in a many-to-many join, where each row of each table can relate to multiple rows of the other table.  The associative entity creates a bridge between the two tables to make the join possible.
